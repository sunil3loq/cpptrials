```python
import pandas as pd
import numpy as np
```

```python
def minmax(x):
    minpred=x[1].min()
    maxpred=x[1].max()
    avepred=x[1].mean()
    binones=(x[0]==1).sum()
    binzeros=(x[0]==0).sum()
    totalvals=x[0].count()
    return pd.Series([minpred,maxpred,avepred,binones,binzeros,totalvals],
    index=['minpred','maxpred','avepred','binones','binzeros','totalvals'])
```

```python
def getcuts(minplist,maxplist,bins):
    '''gets the cuts for each of the bin'''
    cutlist=minplist
    for ind in range(len(minplist)-1):
        cutlist[ind]=(minplist[ind]+maxplist[ind+1])/2
    return pd.Series(cutlist,index=bins)
```

```python
def getksfromcuts(minmaxpred):
    '''from the stage of bin data to KS'''
    minmaxpred['cumones']=minmaxpred['binones'].cumsum()
    minmaxpred['cumzeros']=minmaxpred['binzeros'].cumsum()
    minmaxpred['pctcumones']=minmaxpred['binones'].cumsum()*100/minmaxpred.binones.sum()
    minmaxpred['pctcumzeros']=minmaxpred['binzeros'].cumsum()*100/minmaxpred.binzeros.sum()
    minmaxpred['KS']=minmaxpred['pctcumones']-minmaxpred['pctcumzeros']
    return minmaxpred
```

```python
def makecutsoncol(inppd,colname,numbins=10,cutseries=pd.Series([])):
    '''makes the cuts on the dataframe either quantile manner or from cut scalars'''
    if list(cutseries)==[]:
        inppd['bin']=pd.qcut(inppd[colname],numbins,labels=range(numbins,0,-1))
        minmaxpred=inppd.groupby(['bin']).apply(minmax)
        minmaxpred['maxvaluecuts']=getcuts(list(minmaxpred['minpred']),list(minmaxpred['maxpred']),minmaxpred.index)
    else:
        cutseries.sort()
        binseries=list(cutseries)
        binseries.insert(0,-np.inf)
        binseries[-1]=np.inf
        indexlabels=cutseries.index
        print binseries
        print indexlabels
        inppd['bin']=pd.cut(inppd[colname],bins=binseries,labels=indexlabels)
        minmaxpred=inppd.groupby(['bin']).apply(minmax)
        minmaxpred['maxvaluecuts']=cutseries
    return minmaxpred
```
